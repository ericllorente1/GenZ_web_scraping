```python
import pandas as pd 
```

```python
file_path = 'resources/Web_Search_List_Businesses.csv'
raw_df = pd.read_csv(file_path)
raw_df.head()
raw_df['Fullname'] = raw_df['Leadership Executive(s)First name']+ " " + raw_df['Leadership Executive(s)Last name']
raw_df
```

```python
data = raw_df[['Company name', 'Fullname', 'Leadership Executive(s)Email', 'Street address', 'City', 'Services Offered', 'Employees Statewide']]
email_df = pd.DataFrame(data)
email_df = email_df.fillna('No data')


email_df = email_df.loc[email_df['Leadership Executive(s)Email'] != 'No data']
email_df = email_df.drop_duplicates(subset=["Leadership Executive(s)Email"], keep=False)
email_df = email_df.rename(columns={'Leadership Executive(s)Email': 'Email',
                                   'Street address':'Address'
                                   })
email_df.head()
```

```python
email_df.to_csv('output_files/email_file.csv')
```
