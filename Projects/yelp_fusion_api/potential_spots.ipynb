```python
import pandas as pd
import requests
import json
import pprint
from yelpfusion import api_key
```

```python
import requests
business_type = 'eye care'
business_type = business_type.replace(" ", "%20")
location = 'perth amboy new jersey'
location = location.replace(" ", "%20")
#radius limit is 40000 meters (25 miles)
radius = 30000
url = f"https://api.yelp.com/v3/businesses/search?location={location}&term={business_type}&radius={radius}&sort_by=best_match&limit=50"

headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_key}"
}

response = requests.get(url, headers=headers)
spots = response.json()
businesses = spots['businesses']

```

```python
names = []
cities = []
phones = []
all_address = []
ids = []
websites = []
for i in range(50): 
    name = businesses[i]['name']
    phone = businesses[i]['display_phone']
    city = businesses[i]['location']['city']
    address = businesses[i]['location']['address1']
    id = businesses[i]['id']
    website = businesses[i]['url']
    names.append(name)
    cities.append(city)
    phones.append(phone)
    all_address.append(address)
    ids.append(id)
    websites.append(website)
    
```

```python
potential_spots = pd.DataFrame({'Business Name': names,
                                'Location': cities,
                                'Phone Number': phones,
                               'Address': all_address,
                               'Website': websites})
potential_spots.head(50)
```

```python
potential_spots.to_csv('Output Files/potential_clients_url.csv')
```
